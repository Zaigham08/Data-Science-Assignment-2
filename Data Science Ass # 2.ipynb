```python
import pandas as pd
```

```python
data = pd.read_csv("pokemon.csv")
data.head(n=10)
```

```python
hp_mean = data["HP"].mean()
speed_mean = data["Speed"].mean()

def set_hp(val):
    if val < hp_mean:
        return "Low HP"
    elif val > hp_mean:
        return "High HP"
    else:
        return "Neutral HP"

def set_speed(val):
    if val < speed_mean:
        return "Low Speed"
    elif val > speed_mean:
        return "High Speed"
    else:
        return "Neutral Speed" 

data["High_Low_Speed"] = data["Speed"].apply(set_speed)
data["High_Low_HP"] = data["HP"].apply(set_hp)

data.head(n=12)
```
